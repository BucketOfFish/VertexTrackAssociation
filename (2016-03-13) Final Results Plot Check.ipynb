Functions to perform linear extrapolation for efficiencies.

```python
def triggerRatio(mll, channel):
    if channel==0:
        returnRatio = triggerRatioMMOverEM
    if channel==1:
        returnRatio = triggerRatioEEOverEM
    for bin in range(0,len(triggerMllBins)-1):
        if mll <= triggerMllBins[bin]:
            x = float(mll-triggerMllBins[bin-1])/(triggerMllBins[bin]-triggerMllBins[bin-1])
            return returnRatio[bin-1]+(returnRatio[bin]-returnRatio[bin-1])*x
        if bin==len(triggerMllBins)-2:
            return returnRatio[bin]
        
def offlineRatio(pT, eta):
    if eta<0.1:
        returnRatio = offlineEffMOverE_1
    elif eta<1.1:
        returnRatio = offlineEffMOverE_2
    else:
        returnRatio = offlineEffMOverE_3
    for bin in range(0,len(offlinepTBins)-1):
        if pT <= offlinepTBins[bin]:
            x = float(pT-offlinepTBins[bin-1])/(offlinepTBins[bin]-offlinepTBins[bin-1])
            if offlinepTBins[bin]==27.5:
                return returnRatio[bin]-(returnRatio[bin+1]-returnRatio[bin])*(1-x)
            else:
                return returnRatio[bin-1]+(returnRatio[bin]-returnRatio[bin-1])*x
        if bin==len(offlinepTBins)-2:
            return returnRatio[bin]
```
